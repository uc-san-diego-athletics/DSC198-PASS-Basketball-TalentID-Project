```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('top5bigwest.csv')
```

```python
data
```

```python
teams = data[data['Unnamed: 0'] == 'Team'].reset_index().drop('index', axis=1)
```

```python
teams
```

```python
col = teams.columns.tolist()
col.remove('SCHOOL')
col.remove('Unnamed: 0')
cols = col
fullcol = ['Games', 'Minutes Played', 'Field Goals', 'Field Goals Attempted', 'Field Goal Percentage', '2 Pointers', '2 Pointers Attempted', '2 Point Percentage', '3 Pointers', '3 Pointers Attempted', '3 Point Percentage', 'Free Throws', 'Free Throws Attempted', 'Free Throw Percentage', 'Offensive Rebounds', 'Defensive Rebounds', 'Total Rebounds', 'Assists', 'Steals', 'Blocks', 'Turnovers', 'Personal Fouls', 'Points']
fullcols = {col:fcol for col, fcol in zip(cols, fullcol)}
fullcols
```

```python
for col in cols:
    xaxis = teams['SCHOOL'].unique().tolist()
    yaxis = teams[col].tolist()
    yaxis = [float(i) for i in yaxis]
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.set_title(fullcols[col])
    ax.set_ylim(ymin=0, ymax = max(yaxis) + (max(yaxis) / 5))
    ax.bar(xaxis, yaxis)
    for index,data in enumerate(yaxis):
        plt.text(x=index , y = data - (data/5) , s=f"{data}" , fontdict=dict(fontsize=15), ha='center', c='white')
    plt.show()
    
```

```python
impcols = ['FG%', '2P', '2PA', '2P%', '3P', '3PA', '3P%', 'FT', 'FTA', 'FT%', 'TRB', 'TOV', 'PTS']
for col in impcols:
    xaxis = teams['SCHOOL'].unique().tolist()
    yaxis = teams[col].tolist()
    yaxis = [float(i) for i in yaxis]
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.set_title(fullcols[col])
    ax.set_ylim(ymin=0, ymax = max(yaxis) + (max(yaxis) / 5))
    ax.bar(xaxis, yaxis)
    for index,data in enumerate(yaxis):
        plt.text(x=index , y = data - (data/5) , s=f"{data}" , fontdict=dict(fontsize=15), ha='center', c='white')
    plt.show()
```

```python
data = pd.read_csv('top5bigwest.csv')
ucsdvs = data[(data['Unnamed: 0'] != 'Rank') & (data['SCHOOL'] == 'UCSD')]
ucsdvs
```
